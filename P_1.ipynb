````python
def calculate_water_cost(usage):


    tiers = {
        'Tier 1': {'range': (0, 8000), 'price': 5},
        'Tier 2': {'range': (8001, 22000), 'price': 6},
        'Tier 3': {'range': (22001, 30000), 'price': 7},
        'Tier 4': {'range': (30001, float('inf')), 'price': 10},
    }

    total_cost = 0
    for tier, info in tiers.items():
        if info['range'][0] <= usage <= info['range'][1]:
            tier_usage = usage - info['range'][0] + 1  # Inclusive range
            total_cost += tier_usage * info['price'] / 1000
            break

    return total_cost

# Example usage:
water_usage = 85000  # Gallons
cost = calculate_water_cost(water_usage)
print(f"Total cost for {water_usage} gallons: ${cost:.2f}")
````

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data (replace this with your actual data)
data = {
    'Sample Id': ['S-1', 'S-2', 'S-3', 'S-4', 'S-5', 'S-6', 'S-7', 'S-8', 'S-9', 'S-10'],
    'Species Count': [10, 15, 20, 25, 30, 35, 40, 45, 50, 55]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Function to calculate rarefaction
def rarefaction_curve(sample_count, max_sample_size):
    rarefaction_values = []
    for n in range(1, min(sample_count, max_sample_size) + 1):
        # Calculate the mean of 1000 random samples
        rarefaction_values.append(np.mean([np.random.choice(sample_count, n, replace=False).sum() for _ in range(1000)]))
    return rarefaction_values

# Calculate rarefaction curves for each sample
max_sample_size = df['Species Count'].max()
rarefaction_results = {row['Sample Id']: rarefaction_curve(row['Species Count'], max_sample_size) for index, row in df.iterrows()}

# Plotting
plt.figure(figsize=(12, 6))
colors = sns.color_palette("husl", len(df))  # Generate a color palette

for i, (sample_id, values) in enumerate(rarefaction_results.items()):
    plt.plot(range(1, len(values) + 1), values, label=sample_id, color=colors[i])

plt.title("Rarefaction Curves with Colors")
plt.xlabel("Sample Size")
plt.ylabel("Mean Species Count")
plt.legend(title="Samples")
plt.grid()
plt.show()
````
